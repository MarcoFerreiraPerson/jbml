```python
#was used to download pdfs

button = driver.find_element(By.XPATH, "//a[@data-dt-idx='6']")
button.click()   # click page 180, easier to start at end and work backwards

page_count = 18  # after setting it to 100 on each page, page count will be 18
print(page_count)

scroll_to = driver.find_element(By.XPATH, '//*[@id="dnn_ContentPaneGreyRightColumnsLeft"]/div/div/h2')   # where to scroll to after each page

while (page_count >= 1):
    prev_button = driver.find_element(By.XPATH, "//a[@data-dt-idx='0']")
    prev_button.click()   # click button to go to previous page, the first time it does not matter.
    page_count -= 1
    print(page_count)
    driver.implicitly_wait(5)
    
    driver.execute_script('arguments[0].scrollIntoView(true)', scroll_to)   # scroll to top
    
    Select(driver.find_element(By.XPATH, '//*[@id="data_length"]/label/select')).select_by_value('100')   #select 100 from drop down
    
    # download pdfs
    all_a_tags = driver.find_elements(By.XPATH, "//a")
    count = 0
    for a in all_a_tags:
        if a.get_attribute('href') and '.pdf' in a.get_attribute('href') and 'Download PDF' in a.get_attribute('title'):  
            a.click()
        
            driver.implicitly_wait(2)
            count += 1
            if count >= 100: 
                break
```
